```python
# download 1 class of images
from bing_image_downloader import downloader
downloader.download('preety sunflower',limit=30,output_dir='images',
                    adult_filter_off=True)
```

```python
# download 3rd class of images
downloader.download('ice cream cone',limit=30,output_dir='images',
                    adult_filter_off=True)
```

```python
## data Preprocessing

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
target = []
images = []
flat_data=[]

data_dir = '/content/images'
CATEGORIES = ['preety sunflower','rugby ball leather','ice cream cone']

for category in CATEGORIES:
  class_num = CATEGORIES.index(category)
  print(class_num)
  path=os.path.join(data_dir,category)  # create path to all images
  for img in os.listdir(path):
    img_array= imread(os.path.join(path,img))
    #print(img_array.shape)
    #plt.imshow(img_array)
    img_resized= resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data=np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
 unique,counts= np.unique(target,return_counts=True)
```

```python
plt.bar(CATEGORIES,counts)
```

```python
## train test split
from sklearn.model_selection import train_test_split
xtrain , xtest ,  ytrain  ,ytest = train_test_split(flat_data,target,test_size=0.30,random_state=18)
```

```python
## building model by using SVM

from sklearn.model_selection import GridSearchCV
from sklearn import svm
para_grid = [
            {'C':[1,10,100,1000],'kernel':['linear']},
            {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
             ]

svc = svm.SVC(probability=True)
model=GridSearchCV(svc,para_grid)
model.fit(xtrain,ytrain)
```

```python
yp = model.predict(xtest)
```

```python
yp
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(yp,ytest)
```

```python
confusion_matrix(yp,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators=20)
model_rf.fit(xtrain,ytrain)
```

```python
yp1 = model_rf.predict(xtest)
```

```python
accuracy_score(yp1,ytest)
```

```python
# save model

import pickle
pickle.dump(model_rf,open('image_classification.pkl','wb'))
```
